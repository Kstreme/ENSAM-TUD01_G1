```julia
using QuadGK
using Plots

# Define your function
function original_function(x)
    return cos(x)
end

# Define the fourth primitive function
function fourth_primitive(x)
    # Integrate the original function four times
    result, _ = quadgk(original_function, 0, x)
    result, _ = quadgk(t -> result, 0, x)
    result, _ = quadgk(t -> result, 0, x)
    result, _ = quadgk(t -> result, 0, x)
    
    return result
end

# Create an array of x values
x_values = 0:0.1:10

# Compute the fourth primitive for each x value
fourth_primitive_values = [fourth_primitive(x) for x in x_values]

# Plot the original function and its fourth primitive
plot(x_values, original_function.(x_values), label="Original Function (x^2)", xlabel="x", ylabel="Function Value", legend=:topleft)
plot!(x_values, fourth_primitive_values, label="Fourth Primitive", linestyle=:dash)

```

```julia
using QuadGK
using Plots

# Define your function
function f(x)
    return cos(x)
end

# Define the first three primitives (integrals) of the function
F1(x) = quadgk(f, 0, x)[1]
F2(x) = quadgk(F1, 0, x)[1]
F3(x) = quadgk(F2, 0, x)[1]

# Define the fourth primitive with integration constant set to zero
function F4(x)
    return quadgk(F3, 0, x)[1]
end

# Generate x values for plotting
x_values = -2:0.1:2

# Evaluate the functions at x values
y_f = f.(x_values)
y_F1 = F1.(x_values)
y_F2 = F2.(x_values)
y_F3 = F3.(x_values)
y_F4 = F4.(x_values)

# Plot the original function and its primitives
plot(x_values, [y_f, y_F1, y_F2, y_F3, y_F4], 
    label=["f(x)" "F1(x)" "F2(x)" "F3(x)" "F4(x)"], 
    xlabel="x", ylabel="Function Values",
    linewidth=2)
```

```julia
using QuadGK
using Plots

# Define your function
function f(x)
    return cos(x)
end

# Define the first three primitives (integrals) of the function with initial conditions
F1(x) = quadgk(f, 0, x)[1] - quadgk(f, 0, 0)[1]
F2(x) = quadgk(F1, 0, x)[1] - quadgk(F1, 0, 0)[1]
F3(x) = quadgk(F2, 0, x)[1] - quadgk(F2, 0, 0)[1]

# Define the fourth primitive without the need for a reference point
F4(x) = quadgk(F3, 0, x)[1]

# Generate x values for plotting
x_values = -2:0.1:2

# Evaluate the functions at x values
y_f = f.(x_values)
y_F1 = F1.(x_values)
y_F2 = F2.(x_values)
y_F3 = F3.(x_values)
y_F4 = F4.(x_values)

# Plot the original function and its primitives
plot(x_values, [y_f, y_F1, y_F2, y_F3, y_F4], 
    label=["f(x)" "F1(x)" "F2(x)" "F3(x)" "F4(x)"], 
    xlabel="x", ylabel="Function Values",
    linewidth=2)
```
